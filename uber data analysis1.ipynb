```python
df['month'] = pd.DatetimeIndex(df['START_DATE']).month #It extracts months from start date 
month_label = {1.0 : 'Jan', 2.0 : 'feb', 3.0 : 'march',4.0 : 'April',5.0 : 'may',6.0: 'june',7.0: 'july',8.0: 'August',9.0: 'September',10.0: 'october',11.0: 'November',12.0: 'December'}
df['month'] = df.month.map(month_label) # Number months ko string names se replace karte hai
mon = df.month.value_counts(sort=False) # Mon month ke counts calculate karta hai
```

```python
df.head(12)
```

```python
# Trim any stray spaces in column names
df.rename(columns=lambda c: c.strip(), inplace=True)

# Parse START_DATE to datetime (required for .dt access)
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors='coerce')

# Create/ensure month column (e.g., Jan, Feb, )
if 'month' not in df.columns:
    df['month'] = df['START_DATE'].dt.strftime('%b')

# OPTIONAL: create the 'day' column from START_DATE using your labels
day_label = {0:'Mon',1:'Tues',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}
df['day'] = df['START_DATE'].dt.weekday.map(day_label)
```

```python
grouped = (df.groupby('month', sort=False)['MILES']
             .max()
             .reset_index(name='VALUE COUNT'))

df_plot = grouped.rename(columns={'month':'months'})
```

```python
grouped = (df.groupby('month', sort=False)
             .size()
             .reset_index(name='VALUE COUNT'))

df_plot = grouped.rename(columns={'month':'months'})
```

```python
# Line 1: counts
df_counts = (df.groupby('month', sort=False)
               .size()
               .reset_index(name='Trip Count'))

# Line 2: max miles
df_maxmiles = (df.groupby('month', sort=False)['MILES']
                 .max()
                 .reset_index(name='Max Miles'))

# Merge both into one DataFrame
df_merge = pd.merge(df_counts, df_maxmiles, on="month")

# Plot both
plt.figure(figsize=(12,6))
sns.lineplot(data=df_merge, x="month", y="Trip Count", marker="o", color="blue", label="Trip Count")
sns.lineplot(data=df_merge, x="month", y="Max Miles", marker="o", color="yellow", label="Max Miles")

plt.xlabel("MONTHS")
plt.ylabel("VALUE COUNT")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```
